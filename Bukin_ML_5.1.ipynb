```python
import matplotlib.pyplot as plt

sns.heatmap(res.isnull(), yticklabels=False, cbar=False)

data1.info()

data3_new.info()


import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Function to generate a random date within a specified range
def random_date(start_date, end_date):
    return start_date + timedelta(seconds=random.randint(0, int((end_date - start_date).total_seconds())))

# Creating the first dataset
def create_dataset1(num_entries):
    data = {
        'Id': ['ID_' + str(i) for i in range(1, num_entries + 1)],
        'Channel': [random.choice(['Email', 'Phone', 'Chat']) for _ in range(num_entries)],
        'category': [random.choice(['Technical', 'Billing', 'General']) for _ in range(num_entries)],
        'Sub-category': [random.choice(['Hardware', 'Software', 'Payment Issue']) for _ in range(num_entries)],
        'Issue_reported_Date': [datetime(2023, random.randint(1, 12), random.randint(1, 28)) for _ in range(num_entries)],
        'Issue_responded_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Survey_response_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Agent_name': ['Agent_' + str(random.randint(1, 10)) for _ in range(num_entries)],
        'CSAT Score': [random.randint(1, 5) for _ in range(num_entries)]
    }
    return pd.DataFrame(data)

# Creating the second dataset
def create_dataset2(num_entries):
    data = {
        'Id': ['ID_' + str(i) for i in range(1, num_entries + 1)],
        'Channel': [random.choice(['Email', 'Phone', 'Chat']) for _ in range(num_entries)],
        'category': [random.choice(['Technical', 'Billing', 'General']) for _ in range(num_entries)],
        'Sub-category': [random.choice(['Hardware', 'Software', 'Payment Issue']) for _ in range(num_entries)],
        'Issue_reported_Date': [datetime(2023, random.randint(1, 12), random.randint(1, 28)) for _ in range(num_entries)],
        'Issue_responded_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Survey_response_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Agent_name': ['Agent_' + str(random.randint(1, 10)) for _ in range(num_entries)],
        'CSAT Score': [random.randint(1, 5) for _ in range(num_entries)]
    }
    return pd.DataFrame(data)

# Creating two datasets
dataset1 = create_dataset1(100)
dataset2 = create_dataset2(100)


import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Creating sample data
combined_data_example = pd.concat([dataset1, dataset2], ignore_index=True)

# Creating subplots for categorical data
fig, axs = plt.subplots(3, 2, figsize=(20, 15))

# Bar chart for Channel
combined_data_example['Channel'].value_counts().plot(kind='bar', ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Distribution by Communication Channels')
axs[0, 0].set_xlabel('Channel')
axs[0, 0].set_ylabel('Count')

# Bar chart for category
combined_data_example['category'].value_counts().plot(kind='bar', ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Distribution by Categories')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Count')

# Bar chart for Sub-category
combined_data_example['Sub-category'].value_counts().plot(kind='bar', ax=axs[1, 0], color='lightcoral')
axs[1, 0].set_title('Distribution by Sub-categories')
axs[1, 0].set_xlabel('Sub-category')
axs[1, 0].set_ylabel('Count')

# Visualizing numerical data (CSAT Score)
axs[1, 1].hist(combined_data_example['CSAT Score'], bins=5, color='gold', edgecolor='black')
axs[1, 1].set_title('Distribution of Customer Satisfaction Scores (CSAT)')
axs[1, 1].set_xlabel('Score')
axs[1, 1].set_ylabel('Frequency')

# Visualizing other numerical data
numerical_columns = ['Id', 'Issue_reported_Date', 'Issue_responded_Date', 'Survey_response_Date']
for i, col in enumerate(numerical_columns):
    plt.figure(figsize=(8, 6))
    plt.hist(combined_data_example[col], bins=5, color='lightgrey', edgecolor='black')
    plt.title(col)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

plt.tight_layout()
plt.show()
```

# №5  Изучите документацию pandas в части описания методов merge и join.

## Merge




```python
import pandas as pd

# Создаем два DataFrame для примера
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})

# Объединяем DataFrame по столбцу 'key'
result = pd.merge(left, right, on='key')

print(result)
```

## Join

```python
import pandas as pd

# Создание DataFrame
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]}, index=['a', 'b', 'd'])

# Объединение по индексам
result = df1.join(df2, how='inner') # Тип объединения. Возможные значения: 'left', 'right', 'outer', 'inner'.
print(result)
```

# №6 Познакомьтесь с форматом представления дат strftime.

```python
import datetime

now = datetime.datetime.now()

# Форматирование текущей даты и времени
formatted_date_time = now.strftime("%d-%m-%Y %H:%M:%S")

print("Форматированная дата и время:", formatted_date_time)
```



- `%Y`: 4-значный год (например, 2022)
- `%y`: 2-значный год (например, 22)
- `%m`: Месяц в виде числа (01-12)
- `%B`: Полное название месяца (например, January)
- `%b`: Сокращенное название месяца (например, Jan)
- `%d`: День месяца (01-31)
- `%A`: Полное название дня недели (например, Monday)
- `%a`: Сокращенное название дня недели (например, Mon)
- `%H`: Часы в 24-часовом формате (00-23)
- `%I`: Часы в 12-часовом формате (01-12)
- `%p`: AM или PM
- `%M`: Минуты (00-59)
- `%S`: Секунды (00-59)
- `%f`: Микросекунды
- `%Z`: Название часового пояса (например, UTC, GMT)
- `%z`: Смещение часового пояса в формате ±HHMM или ±HH:MM
