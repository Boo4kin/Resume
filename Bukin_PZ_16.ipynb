```python
C,I,eps,NX= sp.symbols('C,I,epsilon_r,NX')
```

```python
Y, G , r , T = sp.symbols('Y,G,r,T')
```

```python
a0, a1,b0,b1,c0,c1= sp.symbols('a_0, a_1,b_0,b_1,c_0,c_1')
```

```python
sp.init_printing()
```

```python
eql = sp.solve([Y-C-I-G-NX,C-a0-a1*(Y-T),I-b0-b1*r,NX-c0-c1*eps],[C,I,eps,NX],dict=True)[0]
eql
```

```python
dT,dG= sp.symbols('Delta_T,Delta_G')
```

```python
d_eps = sp.simplify(eql[eps].subs(T,T+dT).subs(G,G+dG) - eql[eps])
d_eps
```
