$s$ - доля занятых $E$, которые ежемесячно теряют работу, $s=\text{const}$\
$f$ - доля безработных $U$, которые ежемесяно находят работу, $f=\text{const}$\
$L$ - численность экономически активного населения, $L = \text{const}$\
$U_t$ - численность безработных\
$E_t$ - численность занятых

$\forall\ t$ справедливо $L = U_t + E_t$

$u_t = U_t/L$ - уровень безработицы

Исходя из изложенных предпосылок\
$f \cdot U_t = s \cdot E_t ⇒ f \cdot U_t = s \cdot [L - U_t]\ \ \ / \ \ L$

$f \cdot u_t = s \cdot [1 - u_t] ⇒ u_t = \frac{s}{s + f} = \text{const}$ - естественный уровень безработицы

```python
import sympy as sp
sp.init_printing()
from sympy.series.sequences import RecursiveSeq
```

```python
# s* = y'(k)*k/y(k)

# y(k) = A*k**alpha
```

```python
u= sp.Function('u')
```

```python
s =0.01 ; f=0.2; u0 = 0.068
NUR = s/(s+f) ; NUR
```

```python
t= sp.symbols('t')
```

Справедливо рекуррентное соотношение:\
$U_t = U_{t - 1} - f \cdot U_{t - 1} + s \cdot E_{t - 1}$\
$U_t = (1 - f) \cdot U_{t-1} + s \cdot E_{t-1} \ \ \ / \ \ \ L$\
$u_t = (1 - f) \cdot u_{t-1} + s \cdot (1 - u_{t-1})$\
$u_t = [1 - (s + f)] \cdot u_{t-1} + s$\
Если предел существует, $u_t = u_{t-1}$, тогда \
$u^* = s/(s + f)$

```python
UR = RecursiveSeq((1 - 0.01 - 0.2)*u(t-1)+0.01, u(t), t , [0.068]) ;UR #u0 = 0.068
```

```python
u_List =[] ; d_List= []
for x in zip(range(34),UR):
  u_List.append(x[1])
  d_List.append(2.5*(x[1]-NUR))
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(*[u_List,d_List])),columns= ['u','delta']);df
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 2, figsize = (14, 5))
axs[0].grid(alpha = 0.5)
axs[0].set_xlabel('t')
axs[0].set_ylabel('u')
axs[0].plot(u_List, color = 'blue', lw = 2)
axs[1].grid(alpha = 0.5)
axs[1].set_xlabel('t')
axs[1].set_ylabel('δ')
axs[1].plot(d_List, color = 'red', lw = 2)
```
