# Лаб №1 Метод половинного деления для нахождения корня уравнения

## №1

```python
import numpy as np
import matplotlib.pyplot as plt

# Определим функцию уравнения
def equation(x):
    return x**3 + 2*x + 2

# Создадим массив значений x
x_values = np.linspace(-3, 3, 400)
y_values = equation(x_values)

# Построим график уравнения
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$x^3 + 2x + 2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('График уравнения $x^3 + 2x + 2 = 0$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Определим функцию для метода деления отрезка пополам
def bisection_method(func, a, b, epsilon):
    if func(a) * func(b) > 0:
        print("На этом интервале нет корня")
        return None
    while (b - a) / 2.0 > epsilon:
        midpoint = (a + b) / 2.0
        if func(midpoint) == 0:
            return midpoint
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

# Задаем начальные значения интервала для метода деления отрезка пополам
a = -3
b = 3
epsilon = 0.01

# Применяем метод деления отрезка пополам для нахождения корня
root = bisection_method(equation, a, b, epsilon)
if root is not None:
    print("Уточненное решение методом деления отрезка пополам:", root )
else:
    print("Не удалось найти корень на заданном интервале.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x):
    return x**3 + 2*x +2

# Создание массива значений для построения графика
x_values = np.linspace(-3, 3, 400)
y_values = f(x_values)

# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$x^3 + 4x - 4$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График уравнения $x^3 + 4x - 4 = 0$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Определим производную функции
def derivative(x):
    return x**3+2*x+2

# Метод Ньютона
def newton_method(func, derivative, initial_guess, epsilon):
    x = initial_guess # Начальное приблежение
    iterations = 0
    while abs(func(x)) > epsilon:
        x = x - func(x) / derivative(x)
        iterations += 1
    return x, iterations

# Задаем начальное приближение
initial_guess = 1.5
epsilon = 0.01

# Находим корень методом Ньютона
root_newton, iterations_newton = newton_method(equation, derivative, initial_guess, epsilon)
print("Корень методом Ньютона:", root_newton)
print("Число итераций:", iterations_newton)
```

```python
def chord_method(f, a, b, epsilon):
    iterations = 0
    while abs(b - a) > epsilon:
        a, b = b, b - f(b) * (b - a) / (f(b) - f(a))
        iterations += 1
    return (a + b) / 2, iterations

# Заданные параметры
a = 0
b = 1
epsilon = 0.01

# Вызов метода хорд
root, iterations = chord_method(equation, a, b, epsilon)
print("Метод хорд:")
print("Корень:", root)
print("Число итераций:", iterations)
```

```python
def combined_method(f, f_prime, a, b, epsilon):
    iterations = 0
    while True:
        c = b - f(b) * (b - a) / (f(b) - f(a))
        if abs(c - b) < epsilon:
            break
        a, b = b, c
        b = b - f(b) / f_prime(b)
        iterations += 1
    return b, iterations

# Производная функции
def f_prime(x):
    return -np.sin(x) - 1

# Вызов комбинированного метода
root_combined, iterations_combined = combined_method(equation, f_prime, a, b, epsilon)
print("\nКомбинированный метод:")
print("Корень:", root_combined)
print("Число итераций:", iterations_combined)
```

Якоби:

```python
import numpy as np

# Функция для решения системы линейных уравнений методом Якоби
def my_jacobi_method(A_matrix, b_vector, epsilon, max_iterations):
    size = len(b_vector)
    initial_guess = np.random.rand(size, 1)  # Случайное начальное приближение

    # Итерационный процесс
    for iteration in range(max_iterations):
        # Получаем новое приближение решения, умножая матрицу на предыдущее приближение и добавляя вектор правой части
        solution = np.dot(A_matrix, initial_guess) + b_vector
        # Вычисляем ошибку, как разность между текущим приближением и предыдущим
        error = solution - initial_guess

        # Проверка условия сходимости итерационного процесса
        spectral_radius = np.linalg.norm(A_matrix, ord=1)  # Вычисление спектрального радиуса матрицы
        if spectral_radius >= 1:  # Если спектральный радиус больше или равен 1, итерационный процесс не сходится
            print('Условие сходимости не выполнено')
            return None

        epsilon_i = (spectral_radius / (1 - spectral_radius)) * np.sum(error)  # Вычисление критерия сходимости
        if np.abs(epsilon_i) <= epsilon:  # Если критерий сходимости достигнут
            print(f'На итерации {iteration} с точностью {epsilon} найдены корни:')
            for index in range(size):
                print(f'x{index+1} = {solution[index, 0]}')  # Выводим найденные корни
            return solution

        initial_guess = solution  # Обновляем приближение для следующей итерации

# Матрица коэффициентов и вектор правой части
A = np.array([[0.16, 0.13, -0.32, -0.16],
              [0.34, -0.08, 0.15, -0.18],
              [0.15, 0.32, 0.13, -0.25],
              [0.11, -0.26, -0.08, 0.24]])
b = np.array([-0.64, 1.42, -2.06, 0.83]).reshape(-1, 1)

# Решение методом Якоби
solution = my_jacobi_method(A, b, epsilon=0.001, max_iterations=1000)
```

Зейделя:

```python
import numpy as np

# Функция для решения системы линейных уравнений методом Гаусса-Зейделя
def my_gauss_seidel(A_matrix, b_vector, epsilon):
    size = len(b_vector)
    x = np.zeros(size)  # Инициализация вектора приближенных решений
    iteration = 0  # Счетчик итераций
    converge = False  # Переменная для проверки сходимости
    error = 0.0  # Погрешность

    # Итерационный процесс
    while not converge:
        x_new = np.copy(x)
        for i in range(size):
            # Расчет новых значений переменных по методу Гаусса-Зейделя
            s1 = np.dot(A_matrix[i, :i], x_new[:i])  # Сумма для текущего элемента до диагонали
            s2 = np.dot(A_matrix[i, i + 1:], x[i + 1:])  # Сумма для текущего элемента после диагонали
            x_new[i] = (b_vector[i] - s1 - s2) / A_matrix[i, i]  # Обновление значения переменной

        # Вычисление погрешности
        error = np.linalg.norm(x_new - x, ord=np.inf)
        # Проверка сходимости по достижению требуемой погрешности
        converge = error < epsilon
        iteration += 1  # Увеличение счетчика итераций
        x = x_new  # Обновление значений приближенного решения

    # Вывод результатов
    print('Количество итераций:', iteration)
    print('Решение системы уравнений:', x)
    print('Погрешность:', error)
    return x

# Матрица коэффициентов и вектор правой части
A = np.array([[-0.84, 0.13, -0.32, -0.16],
              [0.34, -1.08, 0.15, -0.18],
              [0.15, 0.32, -0.87, -0.25],
              [0.11, -0.26, -0.08, -0.76]])
b = np.array([0.64, -1.42, 2.06, -0.83])
epsilon = 0.001  # Точность вычислений

# Решение методом Гаусса-Зейделя
x_solution = my_gauss_seidel(A, b, epsilon)
```

Релаксации:

```python
import numpy as np

# Функция решения СЛАУ методом релаксации
def my_relaxation_method(coeff_matrix, right_side, initial_guess, relaxation_parameter, tolerance=0.0001, max_iterations=1000):
    size = len(right_side)
    solution = initial_guess.copy()

    # Итерационный процесс
    for iteration in range(max_iterations):
        new_solution = solution.copy()
        for i in range(size):
            # Применение релаксации для получения нового приближения
            new_solution[i] = (1 - relaxation_parameter) * solution[i] + relaxation_parameter / coeff_matrix[i, i] * \
                              (right_side[i] - np.dot(coeff_matrix[i], solution) + coeff_matrix[i, i] * solution[i])

        # Проверка условия сходимости
        if np.linalg.norm(new_solution - solution) < tolerance:
            return new_solution, iteration+1

        solution = new_solution

    # Если не удалось достичь требуемой точности за максимальное число итераций
    return None, max_iterations

# Функция проверки условия применения метода релаксации
def my_check_diagonal_dominance(coeff_matrix):
    size = coeff_matrix.shape[0]
    for i in range(size):
        # Проверка на строго диагональную доминированность
        if 2 * abs(coeff_matrix[i, i]) <= np.sum(np.abs(coeff_matrix[i])):
            return False
    return True

# Решение СЛАУ методом релаксации
relaxation_parameter = 1.5  # Параметр релаксации
coeff_matrix = np.array([[-0.84, 0.13, -0.32, -0.16],
                         [0.34, -1.08, 0.15, -0.18],
                         [0.15, 0.32, -0.87, -0.25],
                         [0.11, -0.26, -0.08, -0.76]])
right_side = np.array([0.64, -1.42, 2.06, -0.83]) # Свободные члены
initial_guess = np.zeros_like(right_side)
tolerance = 0.001

# Проверка условия применения метода релаксации
if my_check_diagonal_dominance(coeff_matrix):
    print("Условие применения метода релаксации выполняется.")
else:
    raise ValueError("Условие применения метода релаксации не выполняется.")

# Решение системы методом релаксации
solution, num_iterations = my_relaxation_method(coeff_matrix, right_side, initial_guess, relaxation_parameter, tolerance)
if solution is not None:
    print("Решение с точностью ε = 0.001:")
    print(solution)
    print("Количество итераций:", num_iterations)
else:
    print("Метод релаксации не сошелся за", num_iterations, "итераций")
```
