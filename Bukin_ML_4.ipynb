```python
#1
import numpy as np

# Задайте размер таблицы (например, 10x10)
n = 10

# Создайте двумерный массив с единицами
table = np.ones((n, n), dtype=int)

# Заполните массив умножением индексов
for i in range(n):
    for j in range(n):
        table[i, j] = (i + 1) * (j + 1)

# Выведите таблицу умножения
print(table)
```

```python
#2
import numpy as np

def create_arithmetic_matrix(N, first_element, difference):
    # Создайте диагональную матрицу N x N с начальными значениями арифметической прогрессии
    diagonal_values = np.arange(first_element, first_element + N * difference, difference)
    arithmetic_matrix = np.diag(diagonal_values)

    return arithmetic_matrix

# Пример использования функции
N = 5
first_element = 2.0
difference = 3.0

result_matrix = create_arithmetic_matrix(N, first_element, difference)
print(result_matrix)
```

```python
#5
import numpy as np

# Создание матрицы шахматной формы
chessboard_matrix = np.zeros((5, 5), dtype=int)
chessboard_matrix[1::2, ::2] = 5
chessboard_matrix[::2, 1::2] = 5

# Создание треугольной матрицы
triangle_matrix = np.zeros((5, 5), dtype=int)
for i in range(5):
    for j in range(i, 5):
        triangle_matrix[i, j] = 5

# Вычисление детерминантов
det_chessboard = np.linalg.det(chessboard_matrix)
det_triangle = np.linalg.det(triangle_matrix)

# Попытка нахождения обратных матриц
try:
    inv_chessboard = np.linalg.inv(chessboard_matrix)
except np.linalg.LinAlgError as e:
    inv_chessboard = "Невозможно найти обратную матрицу: " + str(e)

try:
    inv_triangle = np.linalg.inv(triangle_matrix)
except np.linalg.LinAlgError as e:
    inv_triangle = "Невозможно найти обратную матрицу: " + str(e)

# Вывод результатов
print("Матрица шахматной формы:")
print(chessboard_matrix)
print("Детерминант матрицы шахматной формы:", det_chessboard)
print("Обратная матрица для матрицы шахматной формы:")
print(inv_chessboard)

print("\nТреугольная матрица:")
print(triangle_matrix)
print("Детерминант треугольной матрицы:", det_triangle)
print("Обратная матрица для треугольной матрицы:")
print(inv_triangle)

```

```python
#17
import pandas as pd
import matplotlib.pyplot as plt

# Загрузите датасет из файла (если он еще не был загружен)
data = pd.read_csv('Data.csv')

# Подсчитайте количество квартир для каждого количества комнат
room_counts = data['Rooms'].value_counts()

# Получите наибольшее количество комнат и выдвиньте сектор
max_rooms = room_counts.idxmax()
explode = [0.1 if rooms == max_rooms else 0 for rooms in room_counts.index]

# Постройте круговую диаграмму
plt.figure(figsize=(8, 8))
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('Распределение квартир по количеству комнат')
plt.axis('equal')  # Это гарантирует, что круг будет круглым
plt.show()
```

```python
#18
import pandas as pd
import matplotlib.pyplot as plt

# Загрузите датасет из файла (если он еще не был загружен)
data = pd.read_csv('Data.csv')

# Постройте гистограмму
plt.figure(figsize=(10, 6))
plt.hist(data['Price'], bins=40, color='black', alpha=0.5)
plt.title('Распределение цен на квартиры')
plt.xlabel('Цена')
plt.ylabel('Количество квартир')
plt.grid(True)

# Определите ценовой интервал с наибольшим количеством квартир
most_common_price_range = data['Price'].mode()[0]
plt.axvline(most_common_price_range, color='green', linestyle='dashed', linewidth=1, label=f'Самый частый ценовой интервал: {most_common_price_range}')
plt.legend()

plt.show()
```

```python
#19
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузите датасет из файла (если он еще не был загружен)
data = pd.read_csv('Data.csv')

# Создайте область figure для всех диаграмм рассеяния
plt.figure(figsize=(16, 12))

# Диаграмма рассеяния для Rooms vs. Price
plt.subplot(2, 2, 1)
sns.scatterplot(x='Rooms', y='Price', data=data)
plt.title('Rooms vs. Price')

# Диаграмма рассеяния для Square vs. Price
plt.subplot(2, 2, 2)
sns.scatterplot(x='Square', y='Price', data=data)
plt.title('Square vs. Price')

# Диаграмма рассеяния для HouseFloor vs. Price
plt.subplot(2, 2, 3)
sns.scatterplot(x='HouseFloor', y='Price', data=data)
plt.title('HouseFloor vs. Price')

# Диаграмма рассеяния для HouseYear vs. Price
plt.subplot(2, 2, 4)
sns.scatterplot(x='HouseYear', y='Price', data=data)
plt.title('HouseYear vs. Price')

# Отобразите графики
plt.tight_layout()
plt.show()
```

```python
#20 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузите датасет из файла (если он еще не был загружен)
data = pd.read_csv('Data.csv')

# Создайте область figure для графиков
plt.figure(figsize=(12, 6))

# Ядерная оценка плотности для Price
plt.subplot(1, 2, 1)
sns.kdeplot(data['Price'], shade=True)
plt.title('Ядерная оценка плотности для Price')

# Двумерная ядерная оценка плотности для Price и HouseFloor
plt.subplot(1, 2, 2)
sns.kdeplot(x=data['Price'], y=data['HouseFloor'], cmap='viridis', fill=True, levels=50)
plt.title('Двумерная KDE для Price и HouseFloor')

# Отобразите графики
plt.tight_layout()
plt.show()


```

```python
#22 
# Выберите только необходимые признаки
features = ['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']
data = data[features]

# Создайте объект PairGrid с различными типами графиков
g = sns.PairGrid(data)

# На диагонали - гистограммы
g.map_diag(sns.histplot, kde_kws={'color': 'k'})

# Под диагональю - ядерные оценки плотности
g.map_lower(sns.kdeplot, cmap='Blues_d')

# Над диагональю - диаграммы рассеяния
g.map_upper(sns.scatterplot)

# Добавьте заголовок к графикам на диагонали
for i, j in zip(*plt.np.triu_indices_from(g.axes, 1)):
    g.axes[i, j].set_visible(False)

# Отобразите графики
plt.show()
```
